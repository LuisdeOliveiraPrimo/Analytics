```python
# Criando uma coluna que comporta a mÃ©dia de preÃ§o de todos os jogos / .round(2) arredonda para dias casas decimais 
df['Average_price'] = ((df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']) / 4).round(2)
df
```

```python
# Dropando uma coluna
df = df.drop('average_price', axis=1)
df
```

```python
import plotly.express as px
import plotly.io as pio

# ForÃ§ar abrir no navegador
pio.renderers.default = 'browser'

# Agrupando os dados
vendas_por_plataforma = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)

# Criando o grÃ¡fico de barras
fig = px.bar(
    vendas_por_plataforma.reset_index(),
    x='Platform',
    y='Global_Sales',
    title='Vendas Globais por Plataforma',
    labels={'Global_Sales': 'Vendas Globais (em milhÃµes)', 'Platform': 'Plataforma'},
    text_auto='.2s'  # Mostra os valores em cima das barras (opcional)
)

# Organizar o eixo X do maior para o menor
fig.update_layout(
    xaxis={'categoryorder': 'total descending'},
    width=800,
    height=500
)

# Mostrar no navegador
fig.show()
```

```python
# Agrupando e selecionando os 10 maiores publishers

import plotly.express as px
import plotly.io as pio

# ForÃ§ar abrir no navegador
pio.renderers.default = 'browser'

# Agrupando dados
vendas_por_publisher = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)

# Criando o grÃ¡fico
fig = px.pie(
    vendas_por_publisher.reset_index(),
    names='Publisher',
    values='Global_Sales',
    title='ParticipaÃ§Ã£o dos 10 Maiores Publishers nas Vendas Globais'
)

# Ajustes de texto no grÃ¡fico
fig.update_traces(textposition='inside', textinfo='percent+label')

# Mostrar
fig.show()
```
